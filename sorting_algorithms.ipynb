# Selection Sort

**Definition:**  
Selection Sort is a simple comparison-based sorting algorithm.  
It repeatedly selects the smallest (or largest) element from the unsorted portion of the list and places it at the correct position in the sorted portion.

---

## Steps of Selection Sort:
1. Start from the first element of the list.
2. Find the smallest element from the unsorted portion.
3. Swap it with the first element of the unsorted portion.
4. Move the boundary between sorted and unsorted parts one step forward.
5. Repeat steps 2–4 until the entire list is sorted.

---

## Example:
Sorting `[29, 10, 14, 37, 13]` in ascending order:

- **Pass 1:** Smallest is `10` → swap with `29` → `[10, 29, 14, 37, 13]`
- **Pass 2:** Smallest from remaining is `13` → swap with `29` → `[10, 13, 14, 37, 29]`
- **Pass 3:** Smallest from remaining is `14` → already in correct position → `[10, 13, 14, 37, 29]`
- **Pass 4:** Smallest from remaining is `29` → swap with `37` → `[10, 13, 14, 29, 37]`

---

## Time Complexity:
- **Best Case:** `O(n²)`
- **Worst Case:** `O(n²)`
- **Average Case:** `O(n²)`

---

## Space Complexity:
- `O(1)` → In-place sorting

---

## Characteristics:
- ❌ **Not adaptive:** Takes the same time even if data is already sorted  
- ✅ **Minimal swaps:** Good when swapping is more costly than comparisons  
- ❌ **Inefficient** for large datasets  
- ✅ Simple to understand and implement  



# Insertion Sort



**Definition:**  
Insertion Sort is a simple sorting algorithm that builds the sorted list one element at a time by inserting each element into its correct position among the previously sorted elements.

---

## Steps of Insertion Sort:
1. Assume the first element is already sorted.
2. Pick the next element from the unsorted portion.
3. Compare it with elements in the sorted portion from right to left.
4. Shift elements in the sorted portion that are greater than the picked element one position to the right.
5. Insert the picked element into its correct position.
6. Repeat until the entire list is sorted.

---

## Example:
Sorting `[5, 3, 4, 1, 2]` in ascending order:

- **Pass 1:** `3` is compared with `5` → insert before `5` → `[3, 5, 4, 1, 2]`
- **Pass 2:** `4` is compared with `5` → insert before `5` → `[3, 4, 5, 1, 2]`
- **Pass 3:** `1` is compared with `5, 4, 3` → insert at start → `[1, 3, 4, 5, 2]`
- **Pass 4:** `2` is compared with `5, 4, 3, 1` → insert after `1` → `[1, 2, 3, 4, 5]`

---

## Time Complexity:
- **Best Case:** `O(n)` → when the list is already sorted
- **Worst Case:** `O(n²)` → when the list is sorted in reverse order
- **Average Case:** `O(n²)`

---

## Space Complexity:
- `O(1)` → In-place sorting (no extra memory required apart from variables)

---

## Characteristics:
- ✅ **Adaptive:** Efficient for small or partially sorted lists  
- ✅ **Stable:** Does not change the order of equal elements  
- ✅ Works well for **small datasets**  
- ✅ Useful when data is continuously received and needs to be sorted on the go
